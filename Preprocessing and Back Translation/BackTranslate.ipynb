```python
import pandas as pd
import string
import csv
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('omw')
nltk.download('omw-1.4')
nltk.download('wordnet')
```

```python
import time
from googletrans import Translator
from nltk.tokenize import word_tokenize

translator = Translator()

def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation.replace('/', ''))
    return text.translate(translator)

def back_translate(src_text, lang_src, lang_dest):
  result = translator.translate(src_text, src=lang_src, dest=lang_dest)
  time.sleep(1)
  back_translated = translator.translate(result.text, dest=lang_src)
  return remove_punctuation(back_translated.text)

def similarity(source, result):
  X_List = word_tokenize(source)
  Y_List = word_tokenize(result)
  l1 =[]
  l2 =[]

  rvector = set(X_List).union(set(Y_List))

  for w in rvector:
    if w in X_List: l1.append(1)
    else: l1.append(0)
    if w in Y_List: l2.append(1)
    else: l2.append(0)
  c = 0

  for i in range(len(rvector)):
    c+= l1[i]*l2[i]
    cosine = c / float((sum(l1)*sum(l2))**0.5)
  return cosine


def translate(data, file_name):
    with open(file_name + '_btranslated.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        for d in data:
            if not d[1]:
                continue
            print("original" + d[1])
            result = back_translate(d[1], "id", "en")
            result = result.lower()
            print("after" + result)
            print(similarity(d[1].lower(), result))
            if similarity(d[1], result) < 0.6:
                writer.writerow([d[0], result, d[2]])
```

```python
data = []
def loadData(file_name):
  global data
  df = pd.read_csv(file_name + '.csv')
  data = df.values.tolist()
```

```python
source = "karena di tempat tinggalnya tidak aman karena banyak kasusnya"
result = "karena di kediamannya tidak aman karena ada banyak kasus"
similarity(source, result)
```

```python
sourcet = "adaptasi dengan lingkungan baru"
print(back_translate(sourcet, 'id', 'en'))
```

```python
def main(file_name):
    loadData(file_name)
    translate(data, file_name)
```

```python
main("data_train_B")
```
